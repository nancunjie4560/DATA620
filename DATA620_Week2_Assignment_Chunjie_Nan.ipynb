Nodes are a basic data structure which contain data and one or more links to other nodes. Nodes can be used to represent a tree structure or a linked list. In such structures where nodes are used, it is possible to traverse from one node to another node.

```python
# import the libraries

import networkx as nx
import matplotlib.pyplot as plt

Graph = nx.Graph()
```



Edge info from krackhardt_kite_graph(), and print the nodes and edges.

```python
kite = nx.generators.small.krackhardt_kite_graph()

print('Nodes:', kite.nodes())
print('Edges:', kite.edges())
print('Number of nodes:', kite.number_of_nodes())
print('Number of edges:', kite.number_of_edges())
```




It shows there are 10 nodes and 18 edges in kite. 




Plot the network.

```python
nx.draw(kite, with_labels = 3)
plt.show()
```
